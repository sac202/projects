```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Customer Call List (1).xlsx")
df
```

```python
df = df.drop(columns = "Not_Useful_Column", errors = "ignore")
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = 
df["Last_Name"] = df["Last_Name"].str.rstrip("123._/")
df
```

```python
#df = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True)


#df["Phone_Number"].apply(lambda x: x[0:3] +'-' +x[3:6]+'-' +x[6:10])
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] +'-' +x[3:6]+'-' +x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '', regex = False)
df
```

```python
df[["Street Address", "States", "Zipcode"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df=df.fillna('')
df
```

```python
df.drop(df[df["Do_Not_Contact"] == 'Y'].index, inplace=True)
df
```

```python
df.drop(df[df["Phone_Number"] == ''].index, inplace=True)
df
```

```python
df = df.reset_index(drop = True)
df
```
